**Matplotlib**:


*   Useful for making plots



```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
# sin wave
plt.plot(x, y)
plt.show()
```

```python
# cosine wave
plt.plot(x, z)
plt.show()
```

```python
# adding title, x-axis & y-axis labels
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('sine value')
plt.title('Sine Wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Parabola')
plt.show()
```

```python
# plotting graph with different colors
plt.plot(x, y, 'r+')
plt.show()
```

```python
plt.plot(x, y, 'g.')
plt.show()
```

```python
plt.plot(x, y, 'rx')
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['Python', 'C++', 'Ruby', 'Java', 'JavaScript']
students = [30, 21, 19, 29, 26]
ax.bar(languages, students)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF STUDENTS')
plt.title('BAR GRAPH')
plt.show()
```

Pie Chart

```python
from enum import auto
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['Python', 'C++', 'Ruby', 'Java', 'JavaScript']
students = [30, 21, 19, 29, 26]
ax.pie(students, labels = languages, autopct = '%1.1f%%')
plt.show()
```

Scatter Plot



*   It is very useful in clustering applications.



```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0, 0, 1, 1])
ax.scatter(x, y, color ='g')
ax.scatter(x, z, color ='b')
plt.show()
```



---

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c = z, cmap = 'Greens')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('3D Grpah')
plt.show()
```
