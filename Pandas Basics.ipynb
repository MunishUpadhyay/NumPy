# Manipulating a DataFrame

**Adding a column to a DataFrame:**

df ['column name'] = dataset_name.column_name

e.g. boston_df ['price'] = boston_dataset.target


**Drop a row:**

boston_df.drop (index = 0, axis = 0)

It will delete the first row with index 0.

It will temporarily delete that row.

**Drop a column:**

boston_df.drop (coulmns = 'column_name', axis = 1)

It will delete the particular column with column_name.

It will temporarily delete that column.

**Locating a row using the index value:**

boston_df.iloc ['row_number']

```python
diabetes_df.iloc[2]
```

```python
#locating a column using the index value
diabetes_df.iloc[:,0]   # prints 1st column
```

```python
diabetes_df.iloc[:,1]   # prints 2nd column
```

```python
diabetes_df.iloc[:,-1]  # prints last column
```
